```python
def freqTable(columnName,dataset):
    freqTable=pd.DataFrame(columns=["unique_values","frequency","relative frequency","cumsum"])
    freqTable["unique_values"]=dataset[columnName].value_counts().index
    freqTable["frequency"]=dataset[columnName].value_counts().values
    freqTable["relative frequency"]=(freqTable["frequency"]/103)
    freqTable["cumsum"]=freqTable["relative frequency"].cumsum()
    return freqTable
```

#frequency is count (no.of times a particular value appers in dataset)
#relative frequency is percentage of times a specific value occurs relative to the total number of observation in dataset(it is 
calculated by frequency value divided by total no of value)
#cummilative frequeny is accumulation of frequency up to certain point by adding relative freqyency
